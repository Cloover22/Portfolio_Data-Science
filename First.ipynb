```python
# EDA
# data 들의 분포를 파악하는 함수
# 
def summarize(data):
    
    # dataframe 의 메모리 사용량을 알아보는 코드이다. datafrmae 은 메모리를 많이 잡아먹기 때문에 너무 크면 문제가 된다.
    size = round(sys.getsizeof(df) / 1024 **2,2)
    
    print(f'data size : {size}MB')
    
    result = pd.DataFrame()
    
    result['Dtype'] = data.dtypes.values # 각 컬럼이 연속형 변수인지, 범주형 변수인지
    result['Count'] = data.count().values # 데이터 행 수
    result['Nunique'] = data.nunique().values # 각 열에 대한 고유한 값의 개수가 몇개나 있는지,고유한 값이 하나면 모두 같은 값이라는 말임
    result['Missing value'] = data.isna().sum().values # 결측치 세기
    result['Most Freq Value'] = data.mode().iloc[0].values # 각 열의 최빈 값을 표현, 최빈 값이 여러개일 경우 첫번재 값을 선택함
    # 각 열의 기초통계값을 표기한다. Transpose 를 사용하는 이유는 describe는 아래와 같은 형태로 표현되어 있기 때문에 통계 값을 컬럼으로 두기 위해서임
    #   A  B  C
# count
# mean
# std
    result['Min'] = data.describe(include = 'all').T['min'].values 
    result['Max'] = data.describe(include = 'all').T['max'].values
    result['Mean'] = data.describe(include = 'all').T['mean'].values
    result['Median'] = data.describe(include = 'all').T['50%'].values
    result = result.set_index(data.columns)
    return result
```
