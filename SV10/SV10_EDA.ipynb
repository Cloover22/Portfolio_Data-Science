```python
# EDA
# data 들의 분포를 파악하는 함수
# 
def summarize(data):
    
    # dataframe 의 메모리 사용량을 알아보는 코드이다. datafrmae 은 메모리를 많이 잡아먹기 때문에 너무 크면 문제가 된다.
    size = round(sys.getsizeof(df) / 1024 **2,2)
    
    print(f'data size : {size}MB')
    
    result = pd.DataFrame()
    
    result['Dtype'] = data.dtypes.values # 각 컬럼이 연속형 변수인지, 범주형 변수인지
    result['Count'] = data.count().values # 데이터 행 수
    result['Nunique'] = data.nunique().values # 각 열에 대한 고유한 값의 개수가 몇개나 있는지,고유한 값이 하나면 모두 같은 값이라는 말임
    result['Missing value'] = data.isna().sum().values # 결측치 세기
    result['Most Freq Value'] = data.mode().iloc[0].values # 각 열의 최빈 값을 표현, 최빈 값이 여러개일 경우 첫번재 값을 선택함
    # 각 열의 기초통계값을 표기한다. Transpose 를 사용하는 이유는 describe는 아래와 같은 형태로 표현되어 있기 때문에 통계 값을 컬럼으로 두기 위해서임
    #   A  B  C
# count
# mean
# std
    result['Min'] = data.describe(include = 'all').T['min'].values 
    result['Max'] = data.describe(include = 'all').T['max'].values
    result['Mean'] = data.describe(include = 'all').T['mean'].values
    result['Median'] = data.describe(include = 'all').T['50%'].values
    result = result.set_index(data.columns)
    return result
```

```python
# 수치형 / 항목형 변수를 나누어 리스트에 저장

numerical_column = []
categorical_column = []

for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
        numerical_column.append(column)
    else:
        categorical_column.append(column)

print('수치형 변수: ',numerical_column)
print('항목형 변수: ',categorical_column)
```

```python
# 수치형 변수 데이터 시각화
for numeric in numerical_column:
    plt.figure(figsize =(5,3))
    plt.hist(df[numeric], bins = 100)
    plt.xlabel(numeric)
    plt.ylabel('frequency')
    plt.title(f'Distribution of {numeric}')
    plt.show()
```

```python
# 항목형 변수 데이터 시각화
for category in categorical_column :
    if category == 'szMessage':
        break
    plt.hist(df[category], bins = 50, edgecolor = 'k')
    plt.xlabel(category)
    plt.ylabel('frequency')
    plt.title(f'Distribution of {category}')
    plt.show()
```

```python
# 각 변수간 상관관계 시각화
correlation_matrix = df[numerical_column].corr()

plt.figure(figsize=(12,10))
sns.heatmap(correlation_matrix, annot = True,fmt = '.1f', cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

각 독립변수간 상관관계를 구했다.
근데 우리가 가지고 있는 컬럼 사이에 상관성이 높은 변수는 정해져있지 않을까?
예를 들어서, DCPA 수치가 낮을수록 위험도는 높아질 것이다. CPA와 관련된 수치들이 작을수록 위험도는 높아질 것이다.

자세한건 결과를 통해서 보자


```python
# 상관성이 비교적 높은 독립변수들을 select
selected_features = correlation_matrix[abs(correlation_matrix) >= 0.8].dropna(how='all', axis=0).dropna(how='all', axis=1)

#  절대값 0.5 이상 상관관계 히트맵 그리기
plt.figure(figsize=(12, 10))
sns.heatmap(selected_features, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Selected Variables)')
plt.show()
```

```python
# 상관관계가 높은 변수들을 리스트에 저장
correlated_pairs = []
for column in selected_features.columns:
    correlated_vars = selected_features.index[selected_features[column].notnull()].tolist()
    correlated_vars.remove(column)
    for var in correlated_vars:
        correlated_pairs.append((column, var))
        

# 상관관계가 높은 변수들 출력
for pair in correlated_pairs:
    print(pair)
```
